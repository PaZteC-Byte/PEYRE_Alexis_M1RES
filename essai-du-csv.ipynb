```python
def type_de_product(): 
    snap = pd.read_csv('mower_market_snapshot.csv', sep=";")
    snap.product_type = pd.Categorical(snap.product_type)
    snap['product_type'] = snap.product_type.cat.codes
    snap['product_type'] = pd.factorize(snap['product_type'])[0] + 1
    print(snap['product_type'])

type_de_product()
```

```python
snap = pd.read_csv('mower_market_snapshot.csv', sep=";")
snap["groupe_1"] = pd.cut(snap.price, [0,300])
snap["groupe_2"] = pd.cut(snap.price, [301,500])
snap["groupe_3"] = pd.cut(snap.price, [501,np.inf])
snap.head()
```

```python
for numbers in snap.price:
        if numbers <= float(300):
            print(numbers, "1")
        elif numbers >= float(301) and numbers <= float(500):
            print(numbers, "2")
        else:
            print(numbers, "3")
```
